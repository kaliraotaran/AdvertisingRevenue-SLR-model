# Here we find the relation ship between the total gain in Revenue as we invest more and more money into Advertisement

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv('advertising_dataset.csv')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
sns.jointplot(x = 'investment', y= 'revenue', data = data)
```

```python

sns.pairplot(data=data)
```

```python
sns.lmplot(x = 'investment', y = 'revenue', data=data)
```

```python
x = data[['investment']]
```

```python
y = data['revenue']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=True)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression(fit_intercept=True)
```

```python
regressor.fit(x_train, y_train)
```

```python
print('Linear Coefficient (m)', regressor.coef_)
print('linear Corfficient (b)', regressor.intercept_)
```

```python
x_test
```

```python
ypredict = regressor.predict(x_test)
ypredict
```

```python
y_test
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, regressor.predict(x_train),color = 'red' )
plt.xlabel('Investment')
plt.ylabel('Revenue')
plt.title('Total gain in Revenue after Investing in Advertisement')
```
